```python

import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
model=load_model('ann_model.h5')
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)
with open('onehot_encoder_geography.pkl','rb') as file:
    onehot_encoder_geography=pickle.load(file)
with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
input_data={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':2,
    'Balance':80000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':90000
}
```

```python
input_df=pd.DataFrame([input_data])
```

```python
new_geo=onehot_encoder_geography.transform([[input_data['Geography']]]).toarray()
geo_df=pd.DataFrame(new_geo,columns=onehot_encoder_geography.get_feature_names_out(['Geography']))
geo_df
```

```python
input_df.drop('Geography',axis=1,inplace=True)
input_df=pd.concat([input_df,geo_df],axis=1)
```

```python
input_df
```

```python
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
##Scaling the input_dtata
input_data_scaled=scaler.transform(input_df)
input_data_scaled
```

```python
prediction=model.predict(input_data_scaled)
prediction
```

```python
if(prediction[0][0]>0.5):
    print("The customer is likely to leave the bank.")
else:
    print("The customer is likely to stay with the bank.")
```
